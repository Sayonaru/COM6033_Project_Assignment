```python
import numpy as np
from scipy import stats

speed = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Calculate Mean
dt_mean = np.mean(speed) ; print ("Mean :",round(dt_mean,2))

# Calculate Median
dt_median = np.median(speed) ; print ("Median :",dt_median)

# Calculate Mode
dt_mode =  stats.mode(speed); print ("Mode :",dt_mode[0])
```

```python
from statistics import variance, stdev
import numpy as np
ages = np.array([5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31])

# Calculate IQR
q75 = np.percentile(ages, [75]); print ("The 75. percentile is",q75)
q90 = np.percentile(ages, [90]); print("90 percent of people are younger than ", q90)
```

```python
# The Python code is as follows:
from scipy import stats
xbar = 990; mu0 = 1000; s = 12.5; n = 30

# Test Statistic
t_smple  = (xbar-mu0)/(s/np.sqrt(float(n))); print ("Test Statistic:",round(t_smple,2))

# Critical value from t-table
alpha = 0.05
t_alpha = stats.t.ppf(alpha,n-1); print ("Critical value from t-table:",round(t_alpha,3))

#Lower tail p-value from t-table
p_val = stats.t.sf(np.abs(t_smple), n-1); print ("Lower tail p-value from t-table", p_val)
```

```python
#The Python code is as follows:
from scipy import stats
xbar = 67; mu0 = 52; s = 16.3

# Calculating z-score
z = (67-52)/16.3

# Calculating probability under the curve
p_val = 1- stats.norm.cdf(z)
print ("Prob. to score more than 67 is ",round(p_val*100,2),"%")
```
